<h3>Imports</h3>

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

<h3>Line Plot</h3>

```python
days = [1,2,3,4,5,6,7]
customerCount = [10, 30, 20, 15, 40, 70, 50]

df = pd.DataFrame({"days":days, "No of customers": customerCount})
df
```

```python
sns.lineplot(x = "days", y = "No of customers", data=df)
plt.show()
```

```python
df = sns.load_dataset("tips")
df.head()
```

```python
sns.lineplot(x = "day", y = "total_bill", data = df, hue="sex")
plt.show()
```

<h3>Bar Plot</h3>

```python
df = sns.load_dataset("titanic")
df.head()
```

```python
sns.barplot(x = "who", y = "fare", data=df)
plt.show()
```

```python
sns.barplot(x = "age", y = "who", data=df, order= ["child", "man", "woman"])
plt.show()
```
